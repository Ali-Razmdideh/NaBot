#merge data

```python
dfMerge = pd.merge(dataMedplus, dataLabtest, on='Lab test')
dfMerge
```

```python
dfMerge.shape
```

```python
dfMerge = pd.merge(dataMedplus, dataLabtest, on='Lab test', how='outer')
dfMerge
```

```python
dfMerge.shape
```

```python
#Lower Casing

dfMerge['Lab test'] = dfMerge['Lab test'].str.lower()
dfMerge['What is it used for'] = dfMerge['What is it used for'].str.lower()
dfMerge['What is the test'] = dfMerge['What is the test'].str.lower()
dfMerge['Why do I need the test'] = dfMerge['Why do I need the test'].str.lower()
dfMerge['What happens during the test?'] = dfMerge['What happens during the test?'].str.lower()
dfMerge['Will I need to do anything to prepare for the test?'] = dfMerge['Will I need to do anything to prepare for the test?'].str.lower()
dfMerge['Are there any risks to the test?'] = dfMerge['Are there any risks to the test?'].str.lower()
dfMerge['What do the results mean?'] = dfMerge['What do the results mean?'].str.lower()
dfMerge['Is there anything else I need to know about the test?'] = dfMerge['Is there anything else I need to know about the test?'].str.lower()
dfMerge['Also Known As'] = dfMerge['Also Known As'].str.lower()
dfMerge['Test Preparation Needed?'] = dfMerge['Test Preparation Needed?'].str.lower()
dfMerge['Why Get Tested?'] = dfMerge['Why Get Tested?'].str.lower()
dfMerge['Sample Required?'] = dfMerge['Sample Required?'].str.lower()
dfMerge['What is being tested?'] = dfMerge['What is being tested?'].str.lower()
dfMerge['When To Get Tested?'] = dfMerge['When To Get Tested?'].str.lower()
dfMerge['When is it ordered?'] = dfMerge['When is it ordered?'].str.lower()
dfMerge['What does the test result mean?'] = dfMerge['What does the test result mean?'].str.lower()
dfMerge['Is there anything else I should know?'] = dfMerge['Is there anything else I should know?'].str.lower()
dfMerge['How is the test used?'] = dfMerge['How is the test used?'].str.lower()
dfMerge['How is it used?'] = dfMerge['How is it used?'].str.lower()
```
