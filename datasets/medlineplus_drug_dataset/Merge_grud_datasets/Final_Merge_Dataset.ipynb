```python
import pandas as pd
df_Medplus = pd.read_csv('drug_dataset_Zahra.csv')
df_Medplus = pd.DataFrame(df_Medplus)
df_Medplus.head(2)
```

```python
# with open('urls.txt', 'r') as file:
#     urls = [line.strip() for line in file]

urls = df_Medplus['title']
print("Tedad Soalat: ",len(set(urls)))
urls = list(urls)
# urls
df_Medplus['title'].value_counts()[:20].tolist()
cols = df_Medplus['title'].value_counts()[:15].index.tolist()
cols
# drugs
```

```python
df11 = pd.DataFrame(columns = ['drug_name']+cols)
df11
```

```python
drugs = set(df_Medplus['drug_name'].unique())
print("Tedad Daroo: ", len(drugs))
drugs = list(drugs)
drugs.sort()
# drugs
#######################

questions = set(df_Medplus['title'].unique())
print("Tedad questions: ", len(questions))
questions = list(questions)
questions.sort()
```

```python
k = 0

for drug in drugs :
    h = 0
    for row in range(len((urls))) :

      if df_Medplus['drug_name'][row] ==  drug :

        for col in cols :
          if df_Medplus['title'][row] ==  col :
            df11.loc[k,['drug_name']] = drug
            df11.loc[k,[col]] = df_Medplus['text'][row]
    k += 1
```
