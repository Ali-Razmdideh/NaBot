```python
import pandas as pd
df_zahra = pd.read_csv('drug_dataset_Zahra.csv')
df_zahra = pd.DataFrame(df_zahra)
df_zahra.head(2)
```

```python
# with open('urls.txt', 'r') as file:
#     urls = [line.strip() for line in file]

urls = df_zahra['title']
print("Tedad Soalat: ",len(set(urls)))
urls = list(urls)
# urls
df_zahra['title'].value_counts()[:20].tolist()
cols = df_zahra['title'].value_counts()[:15].index.tolist()
cols
# drugs
```

```python
df11 = pd.DataFrame(columns = ['drug_name']+cols)
df11
```

```python
drugs = set(df_zahra['drug_name'].unique())
print("Tedad Daroo: ", len(drugs))
drugs = list(drugs)
drugs.sort()
# drugs
#######################

questions = set(df_zahra['title'].unique())
print("Tedad questions: ", len(questions))
questions = list(questions)
questions.sort()
```

```python
k = 0

for drug in drugs :
    h = 0
    for row in range(len((urls))) :

      if df_zahra['drug_name'][row] ==  drug :

        for col in cols :
          if df_zahra['title'][row] ==  col :
            df11.loc[k,['drug_name']] = drug
            df11.loc[k,[col]] = df_zahra['text'][row]
    k += 1
```

```python
df_Zahra = df11
df_Zahra.head(1)
```

```python
# medicine 	vitamin 	dosage 	interactions 	side-effects 	uses 	warnings 	what-to-avoid

df_Zahra.rename(columns={'What side effects can this medication cause?':'side-effects'}, inplace=True)
df_Zahra.rename(columns={'drug_name':'medicine'}, inplace=True)
df_Zahra.rename(columns={'What other information should I know?':'information'}, inplace=True)
df_Zahra.rename(columns={'How should this medicine be used?':'uses'}, inplace=True)
df_Zahra.rename(columns={'Other uses for this medicine':'uses_2'}, inplace=True)
df_Zahra.rename(columns={'IMPORTANT WARNING:':'warnings_2'}, inplace=True)
df_Zahra.rename(columns={'Brand names of combination products':'name_2'}, inplace=True)
df_Zahra.rename(columns={'What should I know about storage and disposal of this medication?':'storage'}, inplace=True)
df_Zahra.rename(columns={'In case of emergency/overdose':'uses_3'}, inplace=True)
df_Zahra.rename(columns={'What should I do if I forget a dose?':'uses_4'}, inplace=True)
df_Zahra.rename(columns={'What special dietary instructions should I follow?':'uses_6'}, inplace=True)
df_Zahra.rename(columns={'What special precautions should I follow?':'uses_6'}, inplace=True)
df_Zahra.rename(columns={'Why is this medication prescribed?':'uses_7'}, inplace=True)
df_Zahra.rename(columns={'IMPORTANT WARNING: ':'warnings'}, inplace=True)


df_Zahra.head(1)
```

```python
df_Zahra.columns
```

```python

df_Zahra['uses'] = df_Zahra['uses'].astype(str) +" -- " + df_Zahra['uses'].astype(str) +" -- " + df_Zahra['uses_2'].astype(str) +" -- " + df_Zahra['uses_3'].astype(str) + " -- " + df_Zahra['uses_4'].astype(str) + " -- " + df_Zahra['uses_5'].astype(str) + " -- " + df_Zahra['uses_6'].astype(str)
# df11.head(1)

df_Zahra['medicine'] = df_Zahra['medicine'].astype(str) +" -- " +  df_Zahra['Brand names'].astype(str) +" -- " + df_Zahra['Other names'].astype(str)+" -- " + df_Zahra['name_2'].astype(str)
df_Zahra.head(1)
```

```python
df_Zahra.drop('uses_2', inplace=True, axis=1)
df_Zahra.drop('uses_3', inplace=True, axis=1)
df_Zahra.drop('uses_4', inplace=True, axis=1)
df_Zahra.drop('uses_5', inplace=True, axis=1)
df_Zahra.drop('uses_6', inplace=True, axis=1)
df_Zahra.drop('uses_7', inplace=True, axis=1)
df_Zahra.drop('Brand names', inplace=True, axis=1)
df_Zahra.drop('IMPORTANT WARNING:', inplace=True, axis=1)
df_Zahra.drop('Other names', inplace=True, axis=1)
df_Zahra.drop('name_2', inplace=True, axis=1)
df_Zahra.drop('Notice:', inplace=True, axis=1)


df_Zahra.head(1)
```
